```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import random
import math
```

```python
data = pd.read_csv('../DSSwithPython-master/DSS-Week08/Data/ACCIDENTS_GU_BCN_2013.csv', encoding='latin-1')
print(data.columns)
```

```python
month = '2013-'+ data['Mes de any'].apply(lambda x: str(x) +'-' )
data['Date'] = month+data['Dia de mes'].apply(lambda x: str(x))
data['Date'] = pd.to_datetime(data['Date'])
accidents = data.groupby('Date').size()
population_mean = accidents.mean()
print ('population mean is ',population_mean)
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,3))
plt.plot(np.arange(0,365), np.array(accidents.values),'b-*',lw=0.7, alpha=0.5)
plt.plot(np.arange(0,365), [population_mean]*365,'r-',lw=0.7, alpha = 0.8)
plt.show()
```

```python
fig1, ax1 = plt.subplots(figsize=(12,4))
plt.ylabel('Frequency')
plt.xlabel('Number of accidents')
plt.hist(accidents.values, bins=20)
ax.axvline(x=accidents.mean(), ymin=0, ymax=40, color=[1, 0, 0])
```
