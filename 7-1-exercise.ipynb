```python
import pulp
import numpy as np
```

```python
model = pulp.LpProblem('manufacturing',pulp.LpMaximize)
p = pulp.LpVariable('p', lowBound=0, upBound=100, cat='continuous')
q = pulp.LpVariable('q', lowBound=0, upBound=40, cat='continuous')
r = pulp.LpVariable('r', lowBound=0, upBound=60, cat='continuous')
```

```python
model += 45*p + 60*q +50*r , 'Profit'
```

```python
model += 20*p + 10*q + 10*r <= 2400
model += 12*p + 28*q + 16*r <= 2400
model += 15*p + 6*q + 16*r <= 2400
model += 10*p + 15*q + 0*r <= 2400
```

```python
model.solve()
```

```python
print('Status:', pulp.LpStatus[model.status])
```

```python
print('optimal profit is obtained when p is', p.varValue)
print('optimal profit is obtained when p is', q.varValue)
print('optimal profit is obtained when p is', r.varValue)
print('max profit is',pulp.value(model.objective))
```

```python
LPP = pulp.LpProblem('Postoffice', pulp.LpMinimize)
x1 = pulp.LpVariable('x1', lowBound=0, cat='Integer')
x2 = pulp.LpVariable('x2', lowBound=0, cat='Integer')
x3 = pulp.LpVariable('x3', lowBound=0, cat='Integer')
x4 = pulp.LpVariable('x4', lowBound=0, cat='Integer')
x5 = pulp.LpVariable('x5', lowBound=0, cat='Integer')
x6 = pulp.LpVariable('x6', lowBound=0, cat='Integer')
x7 = pulp.LpVariable('x7', lowBound=0, cat='Integer')
```

```python
LPP += x1+x2+x3+x4+x5+x6+x7,'Number_of_Employees'
```

```python
LPP += x1 + x4 + x5 + x6 + x7 >= 17
LPP += x2 + x5 + x6 + x7 + x1 >= 13
LPP += x3 + x6 + x7 + x1 + x2 >= 15
LPP += x4 + x7 + x1 + x2 + x3 >= 19
LPP += x5 + x4 + x3 + x2 + x1 >= 14
LPP += x6 + x4 + x5 + x3 + x2 >= 16
LPP += x7 + x4 + x5 + x6 + x3 >= 11
```

```python
LPP.solve()
```

```python
print('status:',pulp.LpStatus[LPP.status])
```

```python
print('x1 is',x1.varValue)
print('x2 is',x2.varValue)
print('x3 is',x3.varValue)
print('x4 is',x4.varValue)
print('x5 is',x5.varValue)
print('x6 is',x6.varValue)
print('x7 is',x7.varValue)
print('Minimun no of employees is:',pulp.value(LPP.objective))
```
