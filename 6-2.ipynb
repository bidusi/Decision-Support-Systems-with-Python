```python
filtered_data = edu[edu['TIME']>2005]
pivedu = filtered_data.pivot_table(index='GEO', columns='TIME', values='Value')
pivedu.head()
```

```python
pivedu.loc[['Spain','Portugal'],[2006,2011]]
```

```python
pivedu
```

```python
pivedu= pivedu.drop(['Euro area (13 countries)',
                      'Euro area (15 countries)',
                      'Euro area (17 countries)',
                      'Euro area (18 countries)',
                      'European Union (25 countries)',
                      'European Union (27 countries)',
                      'European Union (28 countries)'
                      ], axis=0)
```

```python
pivedu = pivedu.rename(index={'Germany (until 1990 former territory of the FRG)':'Germany'})
pivedu = pivedu.dropna()
pivedu.sort_values(by=2011, ascending=False).head()
```

```python
pivedu.rank(ascending=False, method='first').sort_values(by=2011).head()
```

```python
totalSum = pivedu.sum(axis=1)
totalSum.rank(ascending=False, method='first').sort_values().head()
```

```python
pivedu['total'] = pivedu.sum(axis=1)
```

```python
pivedu = pivedu.sort_values(by='total', ascending=False)
pivedu
```

```python
pivedu = pivedu.drop('total', axis=1)
pivedu
```

```python
fig = plt.figure(figsize=(12,5))
totalSum = totalSum.sort_values(ascending=False)
totalSum.plot(kind='bar', alpha=0.5, title='Total Value of Different Countries')
plt.savefig('./Output/Totalvalue_Country.png', dpi=300, bbox_inches = 'tight')
```

```python
fig = plt.figure(figsize=(12,5))
pivedu.plot(kind='barh',stacked=True, alpha = 0.5)
plt.legend(bbox_to_anchor =(1,0.5))
plt.savefig('./Output/Value_time_country.png', dpi=300, bbox_inches = 'tight')
```
